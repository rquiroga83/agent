```python
from langchain_core.tools import tool
import requests


@tool(
    description="Get the weather for a given city."
)
def get_weather(city: str) -> str:
    """Get the weather for a given city."""
    api_key = os.getenv("WEATHER_API_KEY")
    response = requests.get(f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={city}")
    print(response.json())
    return response.json()["current"]["condition"]["text"]



get_weather.invoke({"city": "Bogota"})

```

```python
system_prompt = """Eres un asistente que proporciona informacion sobre el clima en diferentes ciudades.
Cuando un usuario te pregunta sobre el clima en una ciudad, utiliza la herramienta get_weather para obtener
la informacion del clima y proporcionasela al usuario.
"""

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": "Â¿Que clima hay en Bogota?"},
]


llm_tool = llm.bind_tools([get_weather])
response = llm_tool.invoke(messages)
response.pretty_print()

    
```
