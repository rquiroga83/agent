```python
from langchain_deepseek import ChatDeepSeek

llm = ChatDeepSeek(model="deepseek-chat")
response = llm.invoke("Hola soy Andres y mi telefono es 123456789")


response.pretty_print()
```

```python
from pydantic import BaseModel, Field

class ContactInfo(BaseModel):   
    name: str = Field(description="The name of the person")
    phone: str = Field(description="The phone number of the person")
    email: str = Field(description="The email address of the person")
    tone: int = Field(description="The tone of the message, either formal or informal", ge=0, le=100)
    age: int = Field(description="The age of the person")
    sentiment: str = Field(description="The sentiment of the message, either positive, negative, or neutral")


messages = [{"role": "system", "content": "You are a helpful assistant that extracts contact information from messages."},
            {"role": "user", "content": "Hola, mi nombre es Maria, tengo 28 años y mi correo es user123@gmail.com y mi telefono es 987654321 estoy muy incorforme con el servicio que prestan."}]

model_whith_schema = llm.with_structured_output(ContactInfo)
reponses = model_whith_schema.invoke(messages)
reponses
```
