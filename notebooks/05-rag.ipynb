```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import Qdrant
import os

# Load PDF
loader = PyPDFLoader("/Users/andres/Downloads/Mobile_Site_Speed_Playbook.pdf")
documents = loader.load()

# Split into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=200)
chunks = text_splitter.split_documents(documents)

# Initialize embeddings
embeddings = OllamaEmbeddings(model="nomic-embed-text")

# Initialize Qdrant client and upload
qdrant = Qdrant.from_documents(
    chunks,
    embeddings,
    url="http://localhost:6333", # Your Qdrant instance URL
    collection_name="class_collection",
    force_recreate=True # Use with caution, recreates the collection
)

print("PDF documents uploaded to Qdrant successfully!")
```
